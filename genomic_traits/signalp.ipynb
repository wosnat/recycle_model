```python
def _read_(fname):
    df = pd.read_csv(f'{dpath}/signalp/{fname}', sep='\t', skiprows=1)
    df['filename'] = fname.replace('.prediction_results.txt', '').replace('_1','').replace('_2','')
    return df

df = pd.concat([_read_(fname) for fname in os.listdir(f'{dpath}/signalp')], ignore_index=True)
df
    
```

```python
df[['locus_tag', 'product']] = df['# ID'].str.split(' ', n=1, expand=True)
df
```

```python
df[['filename', 'locus_tag', 'product',  'Prediction']].to_csv('trait_gene_annotations/10cc_signalP.csv', index=False)
```
