```python
def _compute_lastday_features(df, groupby_cols, nutrient, biomass_prefix):
    bcol=f'{biomass_prefix}[{nutrient}]'
    lod_lastday_threshold = 5
    refmaxday_df = df.loc[df[bcol].ge(lod_lastday_threshold)].groupby(groupby_cols).day.max()
    #refmaxday_df.rename(columns=dict(day=f'last_day{nutrient}'), inplace=True)
    refmaxday_df.name = f'last_day{nutrient}'
    return refmaxday_df
```

```python
from scipy.signal import find_peaks
def _count_peaks(x):
    peaks, peak_props = find_peaks(x, prominence=5, height=5)
    return len(peaks)
    
def _compute_count_peaks(df, groupby_cols, nutrient, biomass_prefix):
    bcol=f'{biomass_prefix}[{nutrient}]'
    refmaxday_df = df.groupby(groupby_cols)[bcol].apply(_count_peaks)
    refmaxday_df.name = 'peaks'
    return refmaxday_df
```
