```python
from scipy.stats import pearsonr, spearmanr, false_discovery_control
#corr_toremove = 
def _get_corr_matrix(comb_df, params_to_add, pca_cols, threshold=0.3, params_on_cols=False):
    rho = comb_df[params_to_add + pca_cols].corr(method='spearman')    
    pval = comb_df[params_to_add + pca_cols].corr(method=lambda x, y: spearmanr(x, y)[1]) - np.eye(*rho.shape)
    #pvaladj = pd.DataFrame(false_discovery_control(pval, axis=None), index=pval.index, columns=pval.columns)
    p = pval.map(lambda x: ''.join(['*' for t in [.05, .01, .001] if x<=t]).center(3))
    # for c in p.columns:
    #     p[c] = p[c].str.center(3)
    rho_text = rho.copy()
    for c in rho_text.columns:
        rho_text[c] = rho_text[c].apply(lambda x: f"{x: .1f}")        
    rho_text = rho_text  +'\n'+ p #.center() #pad(3, 'both') # apply(lambda x: f'{x:^3}'.format('centered'))
    #rho_text = rho.round(2).astype(str)  +' '+ p #.center() #pad(3, 'both') # apply(lambda x: f'{x:^3}'.format('centered'))
    rho1 = rho[np.abs(rho[pca_cols]).ge(threshold).sum(axis=1).ge(1)]
    rho2 = rho1[rho1.index]
    if params_on_cols:
        cols = [c for c in rho2.columns if c in params_to_add]
        index_list = [c for c in rho2.columns if c not in  params_to_add]
        rho2 = rho2.reindex(columns=cols, index=index_list)
    rho_text = rho_text.reindex(columns=rho2.columns, index=rho2.index)
        
    return (rho2, rho_text)
```

```python
from scipy.stats import pearsonr, spearmanr, false_discovery_control
#corr_toremove =
def _get_corr_tuple(tmpdf, m,c1, c2):
    if tmpdf[c1].isna().sum() == 0:
        correlation, pval = spearmanr(tmpdf[c1], tmpdf[c2])
    else:
        correlation, pval = (np.NaN, np.NaN)
    return (m, c1, c2, correlation, pval)
    
def _get_corr_table(df, params_to_add, pca_cols, models):
    corr_list = list()
    for m in models:
        tmpdf = df.loc[df.model.isin([m])]
        corr_list.extend([ _get_corr_tuple(tmpdf, m, c1, c2)
            for c1 in params_to_add
            for c2 in pca_cols])
    corr_df = pd.DataFrame(corr_list, columns=['model', 'c1', 'c2', 'spearmanr', 'pval'])
    corr_df = corr_df.dropna()
    corr_df['adjpval'] = false_discovery_control(corr_df['pval'])
    corr_df['stars'] = corr_df['adjpval'].map(lambda x: ''.join(['*' for t in [.05, .01, .001] if x<=t]).center(3))
    
    corr_df['spearman_text'] = corr_df['spearmanr'].apply(lambda x: f"{x: .1f}") 
    corr_df.loc[corr_df['spearman_text'].isin(['0.0', '-0.0', ' 0.0']), 'spearman_text'] = '0'
    
    corr_df['annot'] = corr_df['spearman_text']+ '\n' + corr_df['stars']

    return corr_df
              
            
```

```python
def _get_corr_matrix_per_model(comb_df, params_to_add_list, ref_col, only_init_extend=True):
    tdf = comb_df.loc[comb_df.Versatile]
    if only_init_extend:
        tdf = tdf.loc[tdf.Phase.isin(['init', 'extend'])]
    corr_df = _get_corr_table(tdf, params_to_add_list, [ref_col], morder)
    model_corr_data = corr_df.pivot(index='model', columns='c1', values='spearmanr').fillna(0)
    model_corr_annot = corr_df.pivot(index='model', columns='c1', values='annot').fillna('')
    model_corr_data = model_corr_data.reindex(index=morder, columns=params_to_add_list)
    model_corr_annot = model_corr_annot.reindex(index=morder, columns=params_to_add_list)
        
    return model_corr_data, model_corr_annot
```

```python
def _print_clustermap_of_model_correlogram(params_to_add_list, params_to_add_names, ref_col, title, only_init_extend=True):
    if params_to_add_names is None:
        params_to_add_names = params_to_add_list
    model_corr_data, model_corr_annot = _get_corr_matrix_per_model(
        comb_df, params_to_add_list, ref_col,  only_init_extend=only_init_extend)
    sns.set_context('paper')
    annot_kws = {"ha": 'right',"va": 'center'}
    g = sns.heatmap(
            model_corr_data,
            cmap=corr_cmap, vmax=1, vmin=-1,  #row_cluster=False,col_cluster=False,
            annot=model_corr_annot, fmt="s", #figsize=(10,3), 
        lw=0.3, 
            mask= (model_corr_annot==''), square=True,
            # cbar_pos=None, #annot_kws=annot_kws,
        xticklabels=params_to_add_names,
        cbar_kws = dict(# use_gridspec=False,
                        location="bottom", shrink=0.3,),
    )
    # g.ax_heatmap.yaxis.set_ticks_position("left")
    #g.xaxis.tick_top()
    #g.tick_params(axis='x', rotation=90)
    plt.title(title)
    return g
```

```python
def _print_clustermap_of_model_correlogram_t(params_to_add_list, params_to_add_names, ref_col, title, only_init_extend=True):
    if params_to_add_names is None:
        params_to_add_names = params_to_add_list
    model_corr_data, model_corr_annot = _get_corr_matrix_per_model(
        comb_df, params_to_add_list, ref_col,  only_init_extend=only_init_extend)
    model_corr_annot = model_corr_annot.T
    model_corr_data = model_corr_data.T
    model_corr_annot.index= params_to_add_names
    model_corr_data.index= params_to_add_names
    
    sns.set_context('paper')
    annot_kws = {"ha": 'right',"va": 'center'}
    g = sns.heatmap(
            model_corr_data,
            cmap=corr_cmap, vmax=1, vmin=-1,  #row_cluster=False,col_cluster=False,
            annot=model_corr_annot, fmt="s", #figsize=(10,3), 
        lw=0.5, linecolor='#373737',
            mask= (model_corr_annot==''), square=True,
            # cbar_pos=None, #annot_kws=annot_kws,
        yticklabels=params_to_add_names,
        cbar_kws = dict(# use_gridspec=False,
            #            location="bottom", 
            shrink=0.2,),
    )
    # g.ax_heatmap.yaxis.set_ticks_position("left")
    #g.xaxis.tick_top()
    #g.tick_params(axis='x', rotation=90)
    plt.title(title)
    return model_corr_annot, model_corr_data, g
```
